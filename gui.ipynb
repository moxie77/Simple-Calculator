```python
from tkinter import *

calculation = ""


#Creating the function for num
def press(num):
    global calculation
    calculation = calculation + str(num)
    execution.set(calculation)

#Creating the function for equal
def equal_press():
    global calculation
    result = str(eval(calculation))
    execution.set(result)
    calculation =""
    
#Creating the clear function
def clear():
    global calculation
    calculation =""
    execution.set("")
    


    
calculator = Tk()
calculator.title('Moxi Calculator')
calculator.geometry('280x150')

execution = StringVar()


write_to =Entry(calculator,textvariable=execution)
write_to.grid(columnspan=4, ipadx=70)

execution.set('Enter')

button1 =Button(calculator,text='1',command = lambda:press(1),width=5 )
button1.grid(row=2,column=0)
button2 =Button(calculator,text='2',command = lambda:press(2),width=5 )
button2.grid(row=2,column=1)
button3 =Button(calculator,text='3',command = lambda:press(3),width=5 )
button3.grid(row=2,column=2)
button_plus = Button(calculator,text='+',command = lambda:press('+'),width=5)
button_plus.grid(row=2,column=3)

button4 =Button(calculator,text='4',command = lambda:press(4),width=5 )
button4.grid(row=3,column=0)
button5 =Button(calculator,text='5',command = lambda:press(5),width=5 )
button5.grid(row=3,column=1)
button6 =Button(calculator,text='6',command = lambda:press(6),width=5 )
button6.grid(row=3,column=2)
button_minus = Button(calculator,text='-',command =lambda:press('-'),width=5)
button_minus.grid(row=3,column=3)

button7 =Button(calculator,text='7',command = lambda:press(7),width=5 )
button7.grid(row=4,column=0)
button8 =Button(calculator,text='8',command = lambda:press(8),width=5 )
button8.grid(row=4,column=1)
button9 =Button(calculator,text='9',command = lambda:press(6),width=5 )
button9.grid(row=4,column=2)
button_mul = Button(calculator,text='x',command =lambda:press('*'),width=5)
button_mul.grid(row=4,column=3)

button0 = Button(calculator,text='0',command = lambda:press(0),width=5)
button0.grid(row=5,column=0)
button_equal = Button(calculator,text='=',command = equal_press,width=5)
button_equal.grid(row=5,column=1)
button_clear = Button(calculator,text='clear', command =clear,width=5)
button_clear.grid(row=5,column=2) 
button_div = Button(calculator,text='/',command =lambda:press('/'),width=5)
button_div.grid(row=5,column=3)


calculator.mainloop()

```
